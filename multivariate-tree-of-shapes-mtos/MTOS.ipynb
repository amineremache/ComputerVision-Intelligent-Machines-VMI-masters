#I. Introduction and presentation of MToS
##1. MToS: A Tree of Shapes for Multivariate Images

* The topographic map of a gray-level image, also called tree of shapes (ToS), provides a **high-level hierarchical representation of the image contents**. 

* This representation, **invariant** to contrast changes and to contrast inversion, has been proved very useful to achieve many image processing and pattern recognition tasks. 

* Its definition relies on the **total ordering** of pixel values, so this representation **does not exist** for color images, or more generally, **multivariate** images. 

* Common workarounds are not satisfactory and yield many problems. 

* This paper presents a method to build a tree-based representation of multivariate images which features marginally the **same properties** of the gray-level ToS.

* Briefly put, the method relies on the **inclusion relationship** between shapes in the image definition domain.

##2. Implementation of the method


### a) Tree of Shapes (ToS)

```
def comp_tree_tos_2d(image, padding='mean', original_size=True, immersion=True, exterior_vertex=0):

    assert len(image.shape) == 2, "This tree of shapes implementation only supports 2d images."
    immersion = bool(immersion)

    res = hg.cpp._component_tree_tree_of_shapes_image2d(image, padding, original_size, immersion, exterior_vertex)
    tree = res.tree()
    altitudes = res.altitudes()

    if original_size or ((not immersion) and padding == "none"):
        size = image.shape
    else:
        if padding == "none":
            size = (image.shape[0] * 2 - 1, image.shape[1] * 2 - 1)
        else:
            if immersion:
                size = ((image.shape[0] + 2) * 2 - 1, (image.shape[1] + 2) * 2 - 1)
            else:
                size = (image.shape[0] + 2, image.shape[1] + 2)

    g = hg.get_4_adjacency_graph(size)
    hg.CptHierarchy.link(tree, g)

    return tree, altitudes
```

### b) Multivariate Tree of Shapes (MToS)

```
def comp_tree_mtos_2d(image, padding='mean', original_size=True, immersion=True):

    assert len(
        image.shape) == 3, "This multivariate tree of shapes implementation only supports multichannel 2d images."

    ndim = image.shape[2]

    trees = tuple(
        comp_tree_tos_2d(
            image[:, :, k],
            padding,
            original_size=False,
            immersion=immersion)[0]
        for k in range(ndim))
    g = hg.CptHierarchy.get_leaf_graph(trees[0])
    shape = hg.CptGridGraph.get_shape(g)

    depth_map = hg.tree_fusion_depth_map(trees)

    tree, altitudes = comp_tree_tos_2d(np.reshape(depth_map, shape), padding="none",
                                                               original_size=False, immersion=False)

    if original_size and (immersion or padding != "none"):
        deleted_vertices = np.ones((tree.num_leaves(),), dtype=np.bool)
        deleted = np.reshape(deleted_vertices, shape)

        if immersion:
            if padding != "none":
                deleted[2:-2:2, 2:-2:2] = False
            else:
                deleted[0::2, 0::2] = False
        else:
            if padding != "none":
                deleted[1:-1, 1::-1] = False

        all_deleted = hg.accumulate_sequential(tree, deleted_vertices, hg.Accumulators.min)
        shape = (image.shape[0], image.shape[1])
    else:
        all_deleted = np.zeros((tree.num_vertices(),), dtype=np.bool)

    holes = altitudes < altitudes[tree.parents()]

    all_deleted = np.logical_or(all_deleted, holes)

    tree, _ = hg.simplify_tree(tree, all_deleted, process_leaves=True)

    g = hg.get_4_adjacency_graph(shape)
    hg.CptHierarchy.link(tree, g)

    return tree
```

```
def show_saliency(gamma=0.5):
    imshow(1-saliency_khalimsky**gamma, cmap="gray");

interact(show_saliency, gamma= FloatSlider(min=0.01, max=16, continuous_update=False, value=0.5, step=0.05));
```

### b) With gamma and the number of regions as parameters

```
def show_saliency(num_regions=200, gamma=0.5):
    cut = explorer.horizontal_cut_from_num_regions(num_regions)
    cut_altitude = cut.altitude()
    filtered_saliency = np.where(saliency_khalimsky>=cut_altitude, saliency_khalimsky, cut_altitude)
    filtered_saliency = (filtered_saliency - cut_altitude) / (1 - cut_altitude)
    imshow(1-filtered_saliency**gamma, cmap="gray");

interact(show_saliency, 
         num_regions = IntSlider(min=3, max=500, continuous_update=False, value=100, step=1),
         gamma= FloatSlider(min=0.01, max=2, continuous_update=False, value=0.5, step=0.05));
```

## Horizontal cuts

```
# compute average color and random color for nodes
mean_color = hg.attribute_mean_vertex_weights(tree, image)
random_color = np.random.rand(tree.num_vertices(), 3)
```

```
def show_cut(num_regions, fill='Mean Color', show_frontiers=True, mode='At least n regions'):
    cut_nodes = explorer.horizontal_cut_from_num_regions(num_regions, at_least=(mode == 'At least n regions'))
    print("Cut with %d regions at altitude %3.2f"%(len(cut_nodes.nodes()), cut_nodes.altitude()))

    im = cut_nodes.reconstruct_leaf_data(tree, mean_color if fill=="Mean Color" else random_color)

    if show_frontiers:
        cut = cut_nodes.graph_cut(tree)
        cut_khal = hg.graph_4_adjacency_2_khalimsky(graph, cut)

        im_khal = np.zeros([size[0] * 2 - 1, size[1] * 2 - 1, 3], dtype=np.float32)
        im_khal[0::2, 0::2, :] = im
        im_khal[0::2, 1::2, :] = im_khal[0::2, 0:-1:2, :]
        im_khal[1::2, 0::2, :] = im_khal[0:-1:2, 0::2, :]
        im_khal[1::2, 1::2, :] = im_khal[0:-1:2, 0:-1:2, :]

        frontiers = cut_khal != 0
        im_khal[frontiers, :] = 0
        im = im_khal

    imshow(im)

interact(show_cut, 
         num_regions = IntSlider(min=10, max=10000, continuous_update=False, value=20, step=100),
         mode=["At least n regions", "At most n regions"],
         fill=["Mean Color", "Random Color"], 
         show_frontiers=True);
```

#IV. Experimental results



```
def dataset_analysis(images):
  data = []
  for img in images:
    image = imageio.imread(img)
    image = image.astype(np.float32)/255 
    size = image.shape[:2]
    print("Image size: ", size)
    t1 = time.time()
    tree = comp_tree_mtos_2d(image)
    t2 = time.time()
    print("Execution time to construct the MToS : {0:.2f} s".format(t2-t1))
    print("Number of nodes in the MToS: ", tree.num_vertices())
    print("Nbr_nodes ratio/Image_size is : ", tree.num_vertices()/size[0]/size[1])
    # The degree of the tree is the total number of it's children i-e the total number nodes that originate from it.
    print("The degree of the tree is : ", tree.degree(tree.root()))
    print("The height of the tree is : ", hg.attribute_depth(tree).max())
    data.append({"size_img":size,"size_tree":tree.num_vertices(),"ratio":tree.num_vertices()/size[0]/size[1],"degree":tree.degree(tree.root()),"depth":hg.attribute_depth(tree).max(),"time":t2-t1})
  sum_nb_nodes = 0
  sum_ratio = 0
  for item in data:
    sum_nb_nodes += item['size_tree']
    sum_ratio += item['ratio']
  print("\nAverage number of nodes is : ", sum_nb_nodes/len(data))   
  print("\nAverage ratio : ", sum_ratio/len(data)) 
  return data, sum_nb_nodes/len(data), sum_ratio/len(data)
data = dataset_analysis(images)
```
